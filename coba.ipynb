```python
# 3
import pandas as pd

# Membaca file csv ke dalam DataFrame
df = pd.read_csv('data_sampah_jabar.csv')

# Memilih kolom yang diperlukan
df = df[['nama_kabupaten_kota', 'jumlah_produksi_sampah', 'tahun']]

# Menghitung total produksi sampah per tahun menggunakan iterrows
produksi_per_tahun = {}
for index, row in df.iterrows():
    tahun = row['tahun']
    jumlah = row['jumlah_produksi_sampah']
    if tahun not in produksi_per_tahun:
        produksi_per_tahun[tahun] = 0
    produksi_per_tahun[tahun] += jumlah

# Menampilkan hasil
for tahun, total in produksi_per_tahun.items():
    print(f"Total produksi sampah di Jawa Barat pada tahun {tahun}: {total} ton")
```

```python
import pandas as pd

# Membaca file CSV ke dalam DataFrame
df = pd.read_csv('data_sampah_jabar.csv')

# Memilih kolom yang diperlukan
df = df[['nama_kabupaten_kota', 'jumlah_produksi_sampah', 'tahun']]

# Menghitung total produksi sampah per Kota/Kabupaten per tahun menggunakan iterrows
produksi_per_kota_tahun = {}
for index, row in df.iterrows():
    kota = row['nama_kabupaten_kota']
    tahun = row['tahun']
    jumlah = row['jumlah_produksi_sampah']
    if kota not in produksi_per_kota_tahun:
        produksi_per_kota_tahun[kota] = {}
    if tahun not in produksi_per_kota_tahun[kota]:
        produksi_per_kota_tahun[kota][tahun] = 0
    produksi_per_kota_tahun[kota][tahun] += jumlah

# Menampilkan hasil
for kota, tahun_data in produksi_per_kota_tahun.items():
    for tahun, total in tahun_data.items():
        print(f"Total produksi sampah di {kota} pada tahun {tahun}: {total} ton")
```
