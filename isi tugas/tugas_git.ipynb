```python
data_panen = {
    "lokasi1": {
        "nama_lokasi": "Kebun A",
        "hasil_panen": {
            "padi": 1200,
            "jagung": 800,
            "kedelai": 500
        }
    },
    "lokasi2": {
        "nama_lokasi": "Kebun B",
        "hasil_panen": {
            "padi": 1500,
            "jagung": 900,
            "kedelai": 450
        }
    },
    "lokasi3": {
        "nama_lokasi": "Kebun C",
        "hasil_panen": {
            "padi": 1100,
            "jagung": 750,
            "kedelai": 600
        }
    },
    "lokasi4": {
        "nama_lokasi": "Kebun D",
        "hasil_panen": {
            "padi": 1300,
            "jagung": 850,
            "kedelai": 550
        }
    },
    "lokasi5": {
        "nama_lokasi": "Kebun E",
        "hasil_panen": {
            "padi": 1400,
            "jagung": 950,
            "kedelai": 480
        }
    }
}

#1. Tampilkan seluruh data
for lokasi, info in data_panen.items():
    print(f"{info['nama_lokasi']}: {info['hasil_panen']}")

#2. Tampilkan jumlah hasil panen jagung dari lokasi2
jagung_lokasi2 = data_panen["lokasi2"]["hasil_panen"]["jagung"]
print(f"Jumlah hasil panen jagung dari lokasi2: {jagung_lokasi2}")

#3. Tampilkan nama lokasi dari lokasi3
nama_lokasi3 = data_panen["lokasi3"]["nama_lokasi"]
print(f"Nama lokasi dari lokasi3: {nama_lokasi3}")

#4. Masukkan jumlah hasil panen padi dan kedelai ke variabel berbeda
padi_hasil = {}
kedelai_hasil = {}

for lokasi, info in data_panen.items():
    padi_hasil[info["nama_lokasi"]] = info["hasil_panen"]["padi"]
    kedelai_hasil[info["nama_lokasi"]] = info["hasil_panen"]["kedelai"]

print("Jumlah hasil panen padi setiap lokasi:", padi_hasil)
print("Jumlah hasil panen kedelai setiap lokasi:", kedelai_hasil)


#5. Buat percabangan untuk memeriksa kondisi hasil panen
for lokasi, info in data_panen.items():
    padi = info["hasil_panen"]["padi"]
    jagung = info["hasil_panen"]["jagung"]
    
    if padi > 1300 or jagung > 800:
        print(f"{info['nama_lokasi']} memerlukan perhatian lebih lanjut.")
    else:
        print(f"{info['nama_lokasi']} dalam kondisi baik.")
```
